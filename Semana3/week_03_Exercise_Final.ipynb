## Question 12

Consider a bipartite network of N nodes, $N_1$ nodes of type 1 and $N_2$ nodes of type 2 (so that $N_1 + N_2 = N$). What is the maximum number of links in this network?

Answer: 

$N_1 \cdot N_2$, para atingir a quantidade máxima de arestas a rede deve ter $N_1 = N_2$.

Demonstração:

$N_1+N_2=N$  (1)

$N_2 = N - N_1$

$N_1 \cdot (N - N_1)$

$N \cdot N_1 - N_1^2$

Derivando e encontrando o ponto de máximo:

$\frac{d}{dx}N \cdot N_1 - N_1^2 = 0$

$N - 2 \cdot N_1 = 0$

$N = 2 \cdot N_1$ (2)

Substituindo (1) em (2):

$2 \cdot N_1 = N_1 + N_2$

<font color="red">$N_1 = N_2$ </font>

$N_1 + N_2 = N$

$N = 2N_1$

$\frac{N}{2} = N_1 = N_2$

Substtituindo na fórmula original temos:

$L_{max} = N_1 \cdot N_2$

$L_{max} = \frac{N}{2} \cdot \frac{N}{2}$

<font color="red">$L_{max} = \frac{N^2}{4}$</font>

## Question 15

We have seen that Facebook’s network is incredibly sparse. Assume it has approximately 1 billion users, each with 1000 friends on average.

1. Suppose Facebook releases its annual report and it shows that while the number of users in the network has stayed the same, the average number of friends per user has increased. Would this imply that the network density increased, decreased, or stayed the same?
2. Suppose instead that both the number of users and the average number of friends per user doubled. Would this imply the network density increased, decreased, or stayed the same?



Answer:
1. A densidade da rede aumenta. Analisando a fórmulda da densidade, $d = \frac{<k>}{N-1}$, em que $<k>$ é o grau da rede, e $N-1$ é o grau máximo da rede, nota-se que para esse caso, como o número de usuários se mantém o mesmo o denominador dessa equação se mantém constante, entretanto o grau médio da rede aumenta, já que o número médio dos amigos por usuário aumenta. Assim, $<k>$ é aumentado, como consequência aumenta-se a densidade.

2. A densidade da diminui. Note que ao dobrar a quantidade de usuários e amigos, tem-se a seguinte fórmula para a densidade: $d = \frac{2<k>}{2N-1}$. Em que o denominador cresce mais do que o numerador. Como base note que para a densidade se manter constante, a densidade deveria ser $d = \frac{2<k>}{2N-2}$. Dessa forma, a densidade da rede diminui.

## Hands On 01


Often in the context of trees, a node with degree 1 is called a *leaf*. Write a function named `get_leaves` that takes a graph as an argument, loops through the nodes, and returns a list of nodes with degree 1.

```python
# NetworkX
def get_leaves_nx(G):
  leafs = [node for node in G.nodes if G.degree(node) == 1]
  return leafs
# Graph-Tool
def get_leaves_gt(G):
  leafs = [G.vp.name[node] for node in G.vertices() if len(G.get_all_neighbors(node)) == 1]
  return leafs
```

```python
# NetworkX
G = nx.Graph()
G.add_edges_from([
        ('a', 'b'),
        ('a', 'd'),
        ('c', 'd'),
    ])
assert set(get_leaves_nx(G)) == {'c', 'b'}
```

```python
# Graph-Tool
G = gt.Graph(directed=True)
vprop = G.new_vertex_property("string")
orderNodes = ['a', 'b', 'c', 'd']
nodes = [G.add_vertex(n=1) for _ in range(0, len(orderNodes))]
for idx in range(0, len(orderNodes)):
  vprop[nodes[idx]] = orderNodes[idx]
G.vertex_properties["name"] = vprop 
G.add_edge(nodes[0], nodes[1]) #aresta de a para b
G.add_edge(nodes[0], nodes[3]) #aresta de a para d
G.add_edge(nodes[2], nodes[3]) #aresta de c para d
assert set(get_leaves_gt(G)) == {'c', 'b'}
```

## Hands On 02


Write a function max_degree that takes a graph as its argument, and returns a 2-tuple with the name and degree of the node with highest degree.

```python
def createGTGraphAdjlist(adjlist):
  f = open(adjlist, "r")
  #splits each line into a list of strings
  lines = [[str(n) for n in x.split()] for x in f.readlines()]
  #closes the file
  f.close()

  #makes the graph
  G = gt.Graph()
  vprop = G.new_vertex_property("string")
  #lists all the nodes in the graph
  nodes = [G.add_vertex(n=1) for _ in range(0, len(lines))]
  dictNameToIdx = {}
  #name every node in graph and map the name of the node to his index in 'nodes' list 
  for idx in range(0, len(lines)):
    nomeInicialLinha = lines[idx][0]
    dictNameToIdx[nomeInicialLinha] = idx
    vprop[nodes[idx]] = nomeInicialLinha
  G.vertex_properties["name"] = vprop 
  #create all the edges in the graph
  for (idxLinha, line) in enumerate(lines):
      #idxLinha é o index do primeiro nome na linha na lista 'nodes' 
      nodeOfTheFirstName = nodes[idxLinha]
      #for every line creates an edge between the first name and all the others in the same line
      for idxNome in range(1, len(line)):
        nameOfPerson = line[idxNome]
        idxOfPersonInNodes = dictNameToIdx[nameOfPerson]
        G.add_edge(nodeOfTheFirstName, nodes[idxOfPersonInNodes])
  return G
```

```python
# NetworkX
def max_degree_nx(G):
  nodes = G.nodes
  degrees = [G.degree(node) for node in nodes]
  max_dg_idx = max(range(len(degrees)), key=degrees.__getitem__)
  return (list(nodes)[max_dg_idx], degrees[max_dg_idx])
# Graph-Tool
def max_degree_gt(G):
  nodes = [node for node in G.vertices()]
  degrees = [len(G.get_all_neighbors(node)) for node in nodes]
  max_dg_idx = max(range(len(degrees)), key=degrees.__getitem__)
  maxDegreeNode = list(nodes)[max_dg_idx]
  return (G.vp.name[maxDegreeNode], degrees[max_dg_idx])
```

```python
# NetworkX
SG = nx.read_adjlist('friends.adjlist')
assert max_degree_nx(SG) == ('Claire', 4)
```

```python
# Graph-Tool
G = createGTGraphAdjlist('friends.adjlist')
assert max_degree_gt(G) == ('Claire', 4)
```

## Hands on 03


Write a function `mutual_friends` that takes a graph and two nodes as arguments, and returns a list (or set) of nodes that are linked to both given nodes. For example, in the graph `SG` drawn above,

    mutual_friends(SG, 'Alice', 'Claire') == ['Frank']

an empty list or set should be returned in the case where two nodes have no mutual friends, e.g. George and Bob in `SG` drawn above.

```python
# NetworkX
def mutual_friends_nx(G, node_1, node_2):
  friends1 = [x for x in G.neighbors(node_1)]
  friends2 = [x for x in G.neighbors(node_2)]
  return list(set(friends1).intersection(friends2))
# Graph-Tool
def findNodeRef(G, nodeName):
  for nodeRef in G.vertices():
    if G.vp.name[nodeRef] == nodeName: return nodeRef
def mutual_friends_gt(G, node_1, node_2):
  refNode1 = findNodeRef(G, node_1)
  refNode2 = findNodeRef(G, node_2)
  friends1 = [G.vp.name[x] for x in G.get_all_neighbors(refNode1)]
  friends2 = [G.vp.name[x] for x in G.get_all_neighbors(refNode2)]
  return list(set(friends1).intersection(friends2))
```

```python
# NetworkX
SG = nx.read_adjlist('friends.adjlist')
assert mutual_friends_nx(SG, 'Alice', 'Claire') == ['Frank']
assert mutual_friends_nx(SG, 'George', 'Bob') == []
assert sorted(mutual_friends_nx(SG, 'Claire', 'George')) == ['Dennis', 'Frank']
```

```python
# Graph-Tool
G = createGTGraphAdjlist('friends.adjlist')
assert mutual_friends_gt(G, 'Alice', 'Claire') == ['Frank']
assert mutual_friends_gt(G, 'George', 'Bob') == []
assert sorted(mutual_friends_gt(G, 'Claire', 'George')) == ['Dennis', 'Frank']
```
